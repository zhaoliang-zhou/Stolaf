# 2-D Random Walk: Return to Origin

When does a simple 2-D random walk return to the origin? Here, we compute the average number of steps until a 2-D random walk first returns to the origin.

First, import some necessary packages:

```
import matplotlib.pyplot as plt
import random
import numpy
```

Define the possible directions for each step of the walk.

```
dirs = numpy.array([[0,1],[0,-1],[1,0],[-1,0]])
```

Define a function that computes the steps until the random walk returns to the origin. Count the firs time it returns to the origin

```
def stepsToOrigin(maxSteps):
  locations = numpy.zeros((1,2)) # 1 row, 2 columns
  
  # first step
  r = random.randrange(4)
  move = dirs[r]  # direction to move
  locations = numpy.append(locations, [move], axis=0)
  
  # take steps until return to origin                  #locations[-1] is the last of the list 
  while (not numpy.array_equal(locations[-1],[0,0])) and (len(locations) < maxSteps): # avoid an infinite loop?
    r = random.randrange(4)
    move = dirs[r]  # direction to move
    nextloc = locations[-1] + move  # next location
    locations = numpy.append(locations, [nextloc], axis=0)
  
  return len(locations)
```

```
stepsToOrigin(100)
```

```
stepList = [stepsToOrigin(10000) for _ in range(100)]
```

```
plt.hist(stepList);
```

# 3-D Random Walk: Return to Origin

```
dirs3 = numpy.array([[0,0,1],[0,0,-1],[1,0,0],[-1,0,0],[0,1,0],[0,-1,0]])
```

```
import numpy
import random
def stepsToOrigin3(maxSteps):
  locations = numpy.zeros((1,3)) # 1 row, 3 columns
  
  # first step
  r = random.randrange(5)
  move = dirs3[r]  # direction to move
  locations = numpy.append(locations, [move], axis=0)
  
  # take steps until return to origin                  #locations[-1] is the last of the list 
  while (not numpy.array_equal(locations[-1],[0,0,0])) and (len(locations) < maxSteps): # avoid an infinite loop?
    r = random.randrange(5)
    move = dirs3[r]  # direction to move
    nextloc = locations[-1] + move  # next location
    locations = numpy.append(locations, [nextloc], axis=0)
  
  return len(locations)
```
